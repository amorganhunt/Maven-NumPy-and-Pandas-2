# Pie Charts
### Pie charts are used for showing composition with categorical data


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
transactions = pd.read_csv("../data_visualization/transactions.csv", parse_dates=["date"])
store_40 = list(range(40, 50))
```

```python
transactions.head()
```

```python
(
transactions
 .query("store_nbr in @store_40 and date.dt.month in [1, 2, 3]")
 .groupby("store_nbr")
 .agg({"transactions": "sum"})
 .plot.pie(y="transactions")
 ).legend(bbox_to_anchor=(1, 1))
```

```python
(transactions
 .query("store_nbr in @store_40")
 .pivot_table(index=transactions["date"].dt.month,
              columns="store_nbr",
              values="transactions",
              aggfunc="sum")
 .plot.scatter(x=44, y=47)
)
```
