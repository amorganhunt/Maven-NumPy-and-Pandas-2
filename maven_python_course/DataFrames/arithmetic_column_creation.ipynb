# Arithmetic Column Creation
### You can create columns with arithmetic by assigning them Series operations
* Simply specify the new column name and assign the operation of interest

```python
import numpy as np
import pandas as pd
```

```python
# data we'll be working with
oil = pd.read_csv("../DataFrames/oil.csv")
oil['benchmark'] = 90
oil
```

```python
# performing an arithmatic expression on columns to create a new column. and cleaning it up with the round function
oil["benchmark_ratio"] = round((oil.loc[:, "dcoilwtico"] / oil.loc[:, "benchmark"]) * 100, 2)
oil
```

```python
#boolean expression. When the price is at or below our benchmark of 80 we will buy what we can with a budget of 1 million
oil['buy'] = (oil.loc[:, "benchmark_ratio"] < 80) * (1000000 / oil.loc[:, "dcoilwtico"])
oil
```
