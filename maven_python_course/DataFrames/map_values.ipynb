# Mapping Values to Columns
### The .map() method maps values to a column or an entire DataFrame
* You can pass a dictionary with existing values as the keys, and new values as the values

```python
import numpy as np
import pandas as pd
```

```python
retail = pd.read_csv("../DataFrames/retail_2016_2017.csv")

retail.head()
```

```python
# Looking at how may values of each family value exist
retail.loc[:, 'family'].value_counts()
```

```python
# Creating a dictionary that maps the existing values(left) as keys and the new values(right) as values in the new column "Grocery"
product_catagory_dict = {
    'PRODUCE': 'Grocery',
    'POULTRY': 'Grocery',
    'GROCERY I': 'Grocery',
    'GROCERY II': 'Grocery',
    'MEATS': 'Grocery',
    'DELI': 'Grocery',
    'BREAD/BAKERY': 'Grocery',
    'SEAFOOD': 'Grocery',
    'FROZEN FOODS': 'Grocery'
}
```

```python
retail['Grocery'] = retail.loc[:, "family"].map(product_catagory_dict)
```
