```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Step 1: Generate a synthetic dataset
np.random.seed(1)
x = 2.5 * np.random.randn(100) + 1.5   # Array of 100 values with mean = 1.5, stddev = 2.5
res = 0.5 * np.random.randn(100)       # Generate 100 residual terms
y = 2 + 0.3 * x + res                  # Actual values of Y
```

```python
# Step 2: Plot the data to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data points')
plt.title('Synthetic Linear Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 3: Split the data into training and test sets
x = x.reshape(-1, 1)  # Reshape x to be a 2D array
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Step 4: Fit a linear regression model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Step 5: Evaluate the model's performance
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
# Step 6: Plot the regression line and test data
plt.figure(figsize=(10, 6))
plt.scatter(x_test, y_test, color='blue', label='Test data')
plt.plot(x_test, y_pred, color='red', label='Regression line')
plt.title('Linear Regression Analysis')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
